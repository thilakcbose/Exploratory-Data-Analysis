```python
import pandas as pd 

df = pd.read_csv('Employee_Payroll.csv')

df.head()
```

```python
df['Base Pay'] = df['Base Pay'].replace({'\\$': '', ',': ''}, regex=True).astype(float)
df.head()
```

```python
df['Fiscal Year'] = df['Fiscal Period'].str.extract(r'(\d{4})')
df.head()
```

```python
Drop_columns = ['Middle Init', 'Office', 'Job Code', 'Position ID']
df = df.drop(columns=Drop_columns)
df.head()
```

```python
unique_counts = df[['Fiscal Year', 'Office Name', 'Job Title', 'Bureau']].nunique()

print(unique_counts)
```

```python
service_count = df[df['Job Title'].str.contains('Service', case=False)]['Job Title'].nunique()

print(service_count)
```

```python
df['Base Pay'] = df['Base Pay'].replace({'\\$': '', ',': ''}).astype(float)

highest_base_pay = df.groupby('Office Name')['Base Pay'].max()

print(highest_base_pay)
```

```python
new_df = df.set_index('Original Hire Date')[['Job Title', 'Base Pay']]

print(new_df)
```

```python
average_base_pay = df.groupby('Fiscal Year')['Base Pay'].mean()

print(average_base_pay)
```

```python
df['Fiscal Year'] = df['Fiscal Year'].astype(int)

df_2016 = df[df['Fiscal Year'] == 2016]
df_2017 = df[df['Fiscal Year'] == 2017]
df_2018 = df[df['Fiscal Year'] == 2018]


print("DataFrame for 2016:")
print(df_2016.head(1))
print("\nDataFrame for 2017:")
print(df_2017.head(1))
print("\nDataFrame for 2018:")
print(df_2018.head(1))

```

```python
Details = df[(df['Job Title'] == 'Operating Engineer II') & (df['Base Pay'] > 25000) & (df['Base Pay'] < 29000)]

print(Details.head())
```

```python
concatenated_df = pd.concat([df_2016, df_2017, df_2018])

print(concatenated_df)
```

```python
min_base_pay_index = df['Base Pay'].idxmin()

max_base_pay_index = df['Base Pay'].idxmax()

# Extract the rows corresponding to the minimum and maximum 'Base Pay'
min_base_pay_employee = df.loc[min_base_pay_index]
max_base_pay_employee = df.loc[max_base_pay_index]

print("Minimum Base Pay:")
print(min_base_pay_employee)
print("\nMaximum Base Pay:")
print(max_base_pay_employee)
```

```python
highest_base_pay_per_group = df.groupby(['Office Name', 'Job Title'])['Base Pay'].nlargest(2)

print(highest_base_pay_per_group)
```
